````
#criar uma função para retirar colunas conforme nome
def retira_colunas_calculo(textos, dataset):
  """Retira colunas de um dataset com base em parte do nome de uma coluna:
    textos = lista contendo as partes do nome da coluna a pesquisar
    dataset = dataset a ser alterado
    Retorna um dataset sem as colunas que contém os textos escolhidos
  """
  #cria uma lista de colunas a excluir
  colunas = []
  for texto in textos:
    for coluna in dataset.columns:
      if coluna.endswith(texto):
        colunas.append(coluna)
  #exclui as colunas desejadas e retorna um novo dataset
  return dataset.drop(colunas, axis=1)
  
#função para tratamento de nulos
def preenche_tabela(dados):
  """
  Agrupa o dataset pelo identificador da visita do paciente usando a coluna "PATIENT_VISIT_IDENTIFIER", 
  Faz o preenchimento dos dados nulos nas variáveis contínuas usando o forward fill e o backward fill 
  (preenchendo com os dados posteriores primeiro e os anteriores depois)
  Para isso, separa o dataset em 3 partes (features contínuas, features categóricas e saída), faz o
  ajuste nas variáveis categóricas e faz a concatenação após o preenchimento dos nulos em um novo dataset  
  """
  features_continuas_colunas = dados.iloc[:,13:-2].columns
  features_continuas = dados.groupby('PATIENT_VISIT_IDENTIFIER', as_index = False)[features_continuas_colunas].fillna(method='ffill').fillna(method='bfill')
  features_categoricas = dados.iloc[:,:13]
  saida = dados.iloc[:,-2:]
  dados_finais = pd.concat([features_categoricas, features_continuas, saida], ignore_index=True, axis=1 )
  dados_finais.columns = dados.columns
  return dados_finais
#Função para ser usada no apply de um groupby, para colocar em seu escopo que o paciente foi para a UTI
def prepara_janela(linhas):
  """Aplica o valor 1 na coluna ICU se em algum momento o paciente foi para a UTI
   e aplica a primeira janela para cada um dos pacientes
   Pré-Requisitos: Numpy, Pandas
  """
  if(np.any(linhas['ICU'])):
    linhas.loc[linhas['WINDOW'] == '0-2',"ICU"] = 1
  return linhas.loc[linhas['WINDOW'] =='0-2']

#Função para calcular as features usando o RFECV
def selecao_features_por_modelo(X, y, modelo, scorer, min_features_to_select):
  """
  Usa a função RFECV para a escolha de features, de acordo com os dados, o modelo, e a forma de avaliação.
  Retorna o gráfico por feature e um DataFrame com as colunas de X e se a feature é importante para o modelo escolhido

  Variáveis:
  X = Dataframe: São as variáveis explicativas
  y = Dataframe ou Series: A variável alvo
  modelo = um modelo de machine learning de classificação
  scorer = String: métrica de avaliação
            Métricas possíveis  conforme sklearn.metrics.SCORERS.keys(): 
            dict_keys(['explained_variance', 'r2', 
            'max_error', 'neg_median_absolute_error',
             'neg_mean_absolute_error', 'neg_mean_squared_error',
             'neg_mean_squared_log_error', 'neg_root_mean_squared_error',
             'neg_mean_poisson_deviance', 'neg_mean_gamma_deviance', 
             'accuracy', 'roc_auc', 'roc_auc_ovr', 'roc_auc_ovo',
             'roc_auc_ovr_weighted', 'roc_auc_ovo_weighted', 
             'balanced_accuracy', 'average_precision', 'neg_log_loss',
             'neg_brier_score', 'adjusted_rand_score', 'homogeneity_score', 
             'completeness_score', 'v_measure_score', 'mutual_info_score',
             'adjusted_mutual_info_score', 'normalized_mutual_info_score',
             'fowlkes_mallows_score', 'precision', 'precision_macro', 
             'precision_micro', 'precision_samples', 'precision_weighted', 'recall',
             'recall_macro', 'recall_micro', 'recall_samples', 'recall_weighted', 'f1', 
             'f1_macro', 'f1_micro', 'f1_samples', 'f1_weighted', 'jaccard', 
             'jaccard_macro', 'jaccard_micro', 'jaccard_samples', 'jaccard_weighted'])
  min_features_to_select = Int: a quantidade mínima de features a escolher

  Retorno:
  ax = Gráfico contendo a avaliação por quantidade de features
  colunas_a_manter = Dataframe contendo o rfecv.support_ e as colunas de X 


  """
  min_features_to_select = min_features_to_select  # Minimum number of features to consider
  rfecv = RFECV(estimator=modelo, step=1, cv=StratifiedKFold(2),
                scoring= scorer,
              min_features_to_select=min_features_to_select)
  rfecv.fit(X, y)

  print("Optimal number of features : %d" % rfecv.n_features_)

  # Plot number of features VS. cross-validation scores
  plt.figure()
  plt.xlabel("Number of features selected")
  plt.ylabel("Cross validation score (nb of correct classifications)")
  ax = plt.plot(range(min_features_to_select,
               len(rfecv.grid_scores_) + min_features_to_select),
         rfecv.grid_scores_)
  plt.show()

  colunas_a_manter = pd.DataFrame({'coluna':X.columns,
              'bool': rfecv.support_})
  return ax, colunas_a_manter

#Função para plotar correlação
def plota_correlacao(corr, annot):
  """
  Plota um Heatmap de correlação:
  Input:
  corr = Dataframe de correlação: normalmente representado pela função df.corr()
  annot = bool: Se quer que o heatmap venha com os valores ou não
  """
  plt.figure(figsize=(20,20))
  # Criando uma máscara para mostrar apenas os dados únicos do mapa de calor
  mask = np.triu(np.ones_like(corr, dtype=np.bool))
  heatmap = sns.heatmap(corr, mask=mask, vmin=-1, vmax=1, annot=annot, cmap='Blues', fmt = '.2f')

#função que cria um gráfico de feature importance
def cria_grafico_feature_importance(model,data):
  """
  Cria um gráfico com a importância das features num modelo de ensemble.
  Input:
  model: Modelo de Machine Learning Treinado, que contenha entre seus atributos o feature_importances_
  data: os dados usados para ajustar o modelo. (X)

  """
  importances = pd.Series(data=model.feature_importances_, index=data.feature_names)
  sns.barplot(x=importances, y=importances.index, orient='h').set_title('Importância de cada feature')
  

````

````
#FUNÇÃO APENAS CRIANDO AS VARIÁVEIS
def makebio_df_alura(df:pd.DataFrame):

    df["BLOODPRESSURE_ARTERIAL_MEAN"] = (df['BLOODPRESSURE_SISTOLIC_MEAN'] + 2*df['BLOODPRESSURE_DIASTOLIC_MEAN'])/3
 
    df["NEUTROPHILES/LINFOCITOS"] = df['NEUTROPHILES_MEAN']/df['LINFOCITOS_MEAN']

    df["GASO"] = df.groupby("PATIENT_VISIT_IDENTIFIER").P02_ARTERIAL_MEAN.apply(lambda x: x.fillna(method='ffill'))
    df["GASO"] = (~df["GASO"].isna()).astype(int)

    return df
#FUNÇÃO QUE DIVIDE EM DADOS DE TREINO E TESTE USANDO NOSSAS CONFIGURAÇÕES PERSONALIZADAS
def divide_treino_teste(X,y):
  """
  Personaliza a função train_test_split do sklearn
  retorna X_train, X_test, y_train, y_test
  """
  X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.20, random_state=17714, stratify=y)
  return X_train, X_test, y_train, y_test

#FUNÇÃO QUE CRIA E TREINA O MODELO RESULTADO DO TPOT    
def modelo_tpot(X_train,X_test,y_train):
  """
  cria e ajusta um modelo de machine learning com base no resultado de uma AUTOML (TPOT).
  Cria uma pipeline fazendo a padronização dos dados, e aplicando um estimador em Stack, 
  com GradientBoostingClassifier, VarianceThreshold e finalmente o LinearSVC

  Inputs:
   X_train: pd.Dataframe or pd.Series. Variáveis explicativas de treino
   X_test: pd.Dataframe or pd.Series. Variáveis explicativas de teste
   y_train: pd.Series or Array. Variável Alvo do modelo de machine learning para treinar um modelo supervisionado

  Output:
  results ou y_pred = valores preditos pelo modelo 

  """
  # Average CV score on the training set was: 0.775501253132832
  exported_pipeline = make_pipeline(
      StandardScaler(),
      StackingEstimator(estimator=GradientBoostingClassifier(learning_rate=0.1, max_depth=9, 
                                                           max_features=1.0, min_samples_leaf=5, 
                                                           min_samples_split=14, n_estimators=100, 
                                                           subsample=0.45,random_state=17714)),
      VarianceThreshold(threshold=0.05),
      LinearSVC(C=0.001, dual=True, loss="hinge", penalty="l2", tol=0.001, random_state=17714)
  )

  exported_pipeline.fit(X_train, y_train)
  results = exported_pipeline.predict(X_test) 
  return results, exported_pipeline

#função para avaliar o modelo com o classification report e matriz de correlação
def avalia_classificador(X_test, y_test, y_pred, modelo):
  """
  Classifica um modelo treinado, usando o classification report do sci-kit learn
  e plota a matriz de confusão

  Inputs:
  X_test: pd.Dataframe ou pd.Series - variáveis explicativas do dataframe de teste
  y_test: pd.Series ou Array - valores que o modelo pretente prever
  y_pred: pd.Series - previsões do modelo
  modelo: modelo de machine learning treinado com os dados de treino
  """
  print(classification_report(y_test,y_pred))
  plot_confusion_matrix(modelo, X_test, y_test); 

#FUNÇAI QUE CRIA E AVALIA UMA REGRESSÃO LOGÍSTICA DO STATSMODELS
def cria_e_avalia_regressao_logistica(x,y):
  """
  A partir de um conjunto de variáveis explicativas e uma variável alvo, 
  cria um modelo de regressão logística e avalia o mesmo, usando a função summary. 
  Depende do StatsModels. Inclui a inclusão da variável constante, exigida pelo modelo

  Inputs:
  x: pd.Dataframe ou pd.Series - variáveis explicativas do dataframe
  y: pd.Series ou Array - valores que o modelo pretente prever
  """
  x_reglog = sm.add_constant(x)
  # building the model and fitting the data 
  log_reg = sm.Logit(y.astype(float), x_reglog.astype(float)).fit() 
  # printing the summary table 
  print(log_reg.summary()) 

#FUNÇÃO PARA TREINAR E AVALIAR MODELOS
def treina_e_avalia_classificador(modelo, x_treino, x_teste, y_treino, y_teste):
  """
  A partir de um modelo selecionado de classificação, 
  faz o ajuste e a previsão do modelo aos dados de treino, retornando o modelo treinado, 
  as previsões, e a avaliação dos dados de teste usando o classification report e a matriz de confusão.
  Depende da função avalia_classificador():

  Inputs:
  modelo:Objeto = Algoritmo de machine learning. Objeto do tipo modelo (que recebe X e Y como parâmetros, preferencialmente nesta ordem),
  É sugerido que o parâmetros random_state esteja definido entre os hiperparâmetros do modelo, se possível.
  Obrigatoriamente precisa conter os métodos fit() e predict()

  x_treino: pd.DataFrame ou pd.Series = São as variáveis explicativas de treino, para efetuar o ajuste do modelo
  x_teste: pd.Dataframe ou pd.Series - variáveis explicativas do dataframe de teste
  y_treino: pd.Series ou Array - variável alvo usada para treinar o modelo
  y_teste: pd.Series ou Array - valores que o modelo pretente prever

  Outputs:
  modelo_treinado = Um modelo de machine learning treinado com os dados de treino
  previsoes = o y_pred ou as previsões dos dados de teste
  Retorna também a avaliação do modelo
  """
  modelo_treinado = modelo.fit(x_treino, y_treino)
  previsoes = modelo_treinado.predict(x_teste)
  avaliacao = avalia_classificador(x_teste, y_teste, previsoes, modelo_treinado)
  return (modelo_treinado, previsoes,avaliacao)
````

#### **RODANDO UM ALGORITMO DE ESCOLHA DE FEATURES COM BASE NUM MODELO**

De acordo com este [artigo](https://operdata.com.br/blog/analise-de-componentes-principais/):

"*A Análise de Componentes Principais ou PCA (Principal Component Analysis) é uma técnica de análise multivariada que pode ser usada para analisar inter-relações entre um grande número de variáveis e explicar essas variáveis em termos de suas dimensões inerentes (Componentes).*

*O objetivo é encontrar um meio de condensar a informação contida em várias variáveis originais em um conjunto menor de variáveis estatísticas (componentes) com uma perda mínima de informação.*

*O número de componentes principais se torna o número de variáveis consideradas na análise, mas geralmente as primeiras componentes são as mais importantes já que explicam a maior parte da variação total.*

*As componentes principais em geral são extraídas via matriz de covariância, mas também podem ser extraídas via matriz de correlação.* "

Já a Eliminação Recursiva de Atributos (RFE), segundo a documentação do próprio [Scikit-Learn](https://scikit-learn.org/stable/modules/feature_selection.html#rfe):

"*Dado um estimador externo que aplica pesos às variáveis , o objetivo do RFE é selecionar variáveis considerando recursivamente conjuntos cada vez menores de variáveis*

*Na primeira vez que ele roda o estimador usando todas as variáveis possíveis e obtendo a resposta de "importância da variável" de cada atributo, ele retira as variáveis com menos importância, até que o número definido de variáveis seja selecionado*"

Neste caso, usamos a análise de componentes principais para determinar quantas variáveis explicam 95% dos dados, e após usamos este número para escolhermos nossas variáveis.

Vamos fazer o RFE usando dois métodos.
