````
#criar uma funÃ§Ã£o para retirar colunas conforme nome
def retira_colunas_calculo(textos, dataset):
  """Retira colunas de um dataset com base em parte do nome de uma coluna:
    textos = lista contendo as partes do nome da coluna a pesquisar
    dataset = dataset a ser alterado
    Retorna um dataset sem as colunas que contÃ©m os textos escolhidos
  """
  #cria uma lista de colunas a excluir
  colunas = []
  for texto in textos:
    for coluna in dataset.columns:
      if coluna.endswith(texto):
        colunas.append(coluna)
  #exclui as colunas desejadas e retorna um novo dataset
  return dataset.drop(colunas, axis=1)
  
#funÃ§Ã£o para tratamento de nulos
def preenche_tabela(dados):
  """
  Agrupa o dataset pelo identificador da visita do paciente usando a coluna "PATIENT_VISIT_IDENTIFIER", 
  Faz o preenchimento dos dados nulos nas variÃ¡veis contÃ­nuas usando o forward fill e o backward fill 
  (preenchendo com os dados posteriores primeiro e os anteriores depois)
  Para isso, separa o dataset em 3 partes (features contÃ­nuas, features categÃ³ricas e saÃ­da), faz o
  ajuste nas variÃ¡veis categÃ³ricas e faz a concatenaÃ§Ã£o apÃ³s o preenchimento dos nulos em um novo dataset  
  """
  features_continuas_colunas = dados.iloc[:,13:-2].columns
  features_continuas = dados.groupby('PATIENT_VISIT_IDENTIFIER', as_index = False)[features_continuas_colunas].fillna(method='ffill').fillna(method='bfill')
  features_categoricas = dados.iloc[:,:13]
  saida = dados.iloc[:,-2:]
  dados_finais = pd.concat([features_categoricas, features_continuas, saida], ignore_index=True, axis=1 )
  dados_finais.columns = dados.columns
  return dados_finais
#FunÃ§Ã£o para ser usada no apply de um groupby, para colocar em seu escopo que o paciente foi para a UTI
def prepara_janela(linhas):
  """Aplica o valor 1 na coluna ICU se em algum momento o paciente foi para a UTI
   e aplica a primeira janela para cada um dos pacientes
   PrÃ©-Requisitos: Numpy, Pandas
  """
  if(np.any(linhas['ICU'])):
    linhas.loc[linhas['WINDOW'] == '0-2',"ICU"] = 1
  return linhas.loc[linhas['WINDOW'] =='0-2']

#FunÃ§Ã£o para calcular as features usando o RFECV
def selecao_features_por_modelo(X, y, modelo, scorer, min_features_to_select):
  """
  Usa a funÃ§Ã£o RFECV para a escolha de features, de acordo com os dados, o modelo, e a forma de avaliaÃ§Ã£o.
  Retorna o grÃ¡fico por feature e um DataFrame com as colunas de X e se a feature Ã© importante para o modelo escolhido

  VariÃ¡veis:
  X = Dataframe: SÃ£o as variÃ¡veis explicativas
  y = Dataframe ou Series: A variÃ¡vel alvo
  modelo = um modelo de machine learning de classificaÃ§Ã£o
  scorer = String: mÃ©trica de avaliaÃ§Ã£o
            MÃ©tricas possÃ­veis  conforme sklearn.metrics.SCORERS.keys(): 
            dict_keys(['explained_variance', 'r2', 
            'max_error', 'neg_median_absolute_error',
             'neg_mean_absolute_error', 'neg_mean_squared_error',
             'neg_mean_squared_log_error', 'neg_root_mean_squared_error',
             'neg_mean_poisson_deviance', 'neg_mean_gamma_deviance', 
             'accuracy', 'roc_auc', 'roc_auc_ovr', 'roc_auc_ovo',
             'roc_auc_ovr_weighted', 'roc_auc_ovo_weighted', 
             'balanced_accuracy', 'average_precision', 'neg_log_loss',
             'neg_brier_score', 'adjusted_rand_score', 'homogeneity_score', 
             'completeness_score', 'v_measure_score', 'mutual_info_score',
             'adjusted_mutual_info_score', 'normalized_mutual_info_score',
             'fowlkes_mallows_score', 'precision', 'precision_macro', 
             'precision_micro', 'precision_samples', 'precision_weighted', 'recall',
             'recall_macro', 'recall_micro', 'recall_samples', 'recall_weighted', 'f1', 
             'f1_macro', 'f1_micro', 'f1_samples', 'f1_weighted', 'jaccard', 
             'jaccard_macro', 'jaccard_micro', 'jaccard_samples', 'jaccard_weighted'])
  min_features_to_select = Int: a quantidade mÃ­nima de features a escolher

  Retorno:
  ax = GrÃ¡fico contendo a avaliaÃ§Ã£o por quantidade de features
  colunas_a_manter = Dataframe contendo o rfecv.support_ e as colunas de X 


  """
  min_features_to_select = min_features_to_select  # Minimum number of features to consider
  rfecv = RFECV(estimator=modelo, step=1, cv=StratifiedKFold(2),
                scoring= scorer,
              min_features_to_select=min_features_to_select)
  rfecv.fit(X, y)

  print("Optimal number of features : %d" % rfecv.n_features_)

  # Plot number of features VS. cross-validation scores
  plt.figure()
  plt.xlabel("Number of features selected")
  plt.ylabel("Cross validation score (nb of correct classifications)")
  ax = plt.plot(range(min_features_to_select,
               len(rfecv.grid_scores_) + min_features_to_select),
         rfecv.grid_scores_)
  plt.show()

  colunas_a_manter = pd.DataFrame({'coluna':X.columns,
              'bool': rfecv.support_})
  return ax, colunas_a_manter

#FunÃ§Ã£o para plotar correlaÃ§Ã£o
def plota_correlacao(corr, annot):
  """
  Plota um Heatmap de correlaÃ§Ã£o:
  Input:
  corr = Dataframe de correlaÃ§Ã£o: normalmente representado pela funÃ§Ã£o df.corr()
  annot = bool: Se quer que o heatmap venha com os valores ou nÃ£o
  """
  plt.figure(figsize=(20,20))
  # Criando uma mÃ¡scara para mostrar apenas os dados Ãºnicos do mapa de calor
  mask = np.triu(np.ones_like(corr, dtype=np.bool))
  heatmap = sns.heatmap(corr, mask=mask, vmin=-1, vmax=1, annot=annot, cmap='Blues', fmt = '.2f')

#funÃ§Ã£o que cria um grÃ¡fico de feature importance
def cria_grafico_feature_importance(model,data):
  """
  Cria um grÃ¡fico com a importÃ¢ncia das features num modelo de ensemble.
  Input:
  model: Modelo de Machine Learning Treinado, que contenha entre seus atributos o feature_importances_
  data: os dados usados para ajustar o modelo. (X)

  """
  importances = pd.Series(data=model.feature_importances_, index=data.feature_names)
  sns.barplot(x=importances, y=importances.index, orient='h').set_title('ImportÃ¢ncia de cada feature')
  

````

#### **RODANDO UM ALGORITMO DE ESCOLHA DE FEATURES COM BASE NUM MODELO**

De acordo com este [artigo](https://operdata.com.br/blog/analise-de-componentes-principais/):

"*A AnÃ¡lise de Componentes Principais ou PCA (Principal Component Analysis) Ã© uma tÃ©cnica de anÃ¡lise multivariada que pode ser usada para analisar inter-relaÃ§Ãµes entre um grande nÃºmero de variÃ¡veis e explicar essas variÃ¡veis em termos de suas dimensÃµes inerentes (Componentes).*

*O objetivo Ã© encontrar um meio de condensar a informaÃ§Ã£o contida em vÃ¡rias variÃ¡veis originais em um conjunto menor de variÃ¡veis estatÃ­sticas (componentes) com uma perda mÃ­nima de informaÃ§Ã£o.*

*O nÃºmero de componentes principais se torna o nÃºmero de variÃ¡veis consideradas na anÃ¡lise, mas geralmente as primeiras componentes sÃ£o as mais importantes jÃ¡ que explicam a maior parte da variaÃ§Ã£o total.*

*As componentes principais em geral sÃ£o extraÃ­das via matriz de covariÃ¢ncia, mas tambÃ©m podem ser extraÃ­das via matriz de correlaÃ§Ã£o.* "

JÃ¡ a EliminaÃ§Ã£o Recursiva de Atributos (RFE), segundo a documentaÃ§Ã£o do prÃ³prio [Scikit-Learn](https://scikit-learn.org/stable/modules/feature_selection.html#rfe):

"*Dado um estimador externo que aplica pesos Ã s variÃ¡veis , o objetivo do RFE Ã© selecionar variÃ¡veis considerando recursivamente conjuntos cada vez menores de variÃ¡veis*

*Na primeira vez que ele roda o estimador usando todas as variÃ¡veis possÃ­veis e obtendo a resposta de "importÃ¢ncia da variÃ¡vel" de cada atributo, ele retira as variÃ¡veis com menos importÃ¢ncia, atÃ© que o nÃºmero definido de variÃ¡veis seja selecionado*"

Neste caso, usamos a anÃ¡lise de componentes principais para determinar quantas variÃ¡veis explicam 95% dos dados, e apÃ³s usamos este nÃºmero para escolhermos nossas variÃ¡veis.

Vamos fazer o RFE usando dois mÃ©todos. Um com ValidaÃ§Ã£o Cruzada, para que ele mesmo escolha a quantidade de features usando o mÃ©todo StratifiedKFold, seguindo as sugestÃµes deste artigo do [Medium](https://medium.com/data-hackers/como-selecionar-as-melhores-features-para-seu-modelo-de-machine-learning-faf74e357913)
