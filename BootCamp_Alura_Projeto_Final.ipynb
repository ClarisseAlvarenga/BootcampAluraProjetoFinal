### **FUNÇÕES**

```
#criar uma função para retirar colunas conforme nome
def retira_colunas_calculo(textos, dataset):
  """Retira colunas de um dataset com base em parte do nome de uma coluna:
    textos = lista contendo as partes do nome da coluna a pesquisar
    dataset = dataset a ser alterado
    Retorna um dataset sem as colunas que contém os textos escolhidos
  """
  #cria uma lista de colunas a excluir
  colunas = []
  for texto in textos:
    for coluna in dataset.columns:
      if coluna.endswith(texto):
        colunas.append(coluna)
  #exclui as colunas desejadas e retorna um novo dataset
  return dataset.drop(colunas, axis=1)
#função para tratamento de nulos
def preenche_tabela(dados):
  """
  Agrupa o dataset pelo identificador da visita do paciente usando a coluna "PATIENT_VISIT_IDENTIFIER", 
  Faz o preenchimento dos dados nulos nas variáveis contínuas usando o forward fill e o backward fill 
  (preenchendo com os dados posteriores primeiro e os anteriores depois)
  Para isso, separa o dataset em 3 partes (features contínuas, features categóricas e saída), faz o
  ajuste nas variáveis categóricas e faz a concatenação após o preenchimento dos nulos em um novo dataset  
  """
  features_continuas_colunas = dados.iloc[:,13:-2].columns
  features_continuas = dados.groupby('PATIENT_VISIT_IDENTIFIER', as_index = False)[features_continuas_colunas].fillna(method='ffill').fillna(method='bfill')
  features_categoricas = dados.iloc[:,:13]
  saida = dados.iloc[:,-2:]
  dados_finais = pd.concat([features_categoricas, features_continuas, saida], ignore_index=True, axis=1 )
  dados_finais.columns = dados.columns
  return dados_finais
#Função para ser usada no apply de um groupby, para colocar em seu escopo que o paciente foi para a UTI
def prepara_janela(linhas):
  """Aplica o valor 1 na coluna ICU se em algum momento o paciente foi para a UTI
   e aplica a primeira janela para cada um dos pacientes
   Pré-Requisitos: Numpy, Pandas
  """
  if(np.any(linhas['ICU'])):
    linhas.loc[linhas['WINDOW'] == '0-2',"ICU"] = 1
  return linhas.loc[linhas['WINDOW'] =='0-2']

#Função para calcular as features usando o RFECV
def selecao_features_por_modelo(X, y, modelo, scorer, min_features_to_select):
  """
  Usa a função RFECV para a escolha de features, de acordo com os dados, o modelo, e a forma de avaliação.
  Retorna o gráfico por feature e um DataFrame com as colunas de X e se a feature é importante para o modelo escolhido

  Variáveis:
  X = Dataframe: São as variáveis explicativas
  y = Dataframe ou Series: A variável alvo
  modelo = um modelo de machine learning de classificação
  scorer = String: métrica de avaliação
            Métricas possíveis  conforme sklearn.metrics.SCORERS.keys(): 
            dict_keys(['explained_variance', 'r2', 
            'max_error', 'neg_median_absolute_error',
             'neg_mean_absolute_error', 'neg_mean_squared_error',
             'neg_mean_squared_log_error', 'neg_root_mean_squared_error',
             'neg_mean_poisson_deviance', 'neg_mean_gamma_deviance', 
             'accuracy', 'roc_auc', 'roc_auc_ovr', 'roc_auc_ovo',
             'roc_auc_ovr_weighted', 'roc_auc_ovo_weighted', 
             'balanced_accuracy', 'average_precision', 'neg_log_loss',
             'neg_brier_score', 'adjusted_rand_score', 'homogeneity_score', 
             'completeness_score', 'v_measure_score', 'mutual_info_score',
             'adjusted_mutual_info_score', 'normalized_mutual_info_score',
             'fowlkes_mallows_score', 'precision', 'precision_macro', 
             'precision_micro', 'precision_samples', 'precision_weighted', 'recall',
             'recall_macro', 'recall_micro', 'recall_samples', 'recall_weighted', 'f1', 
             'f1_macro', 'f1_micro', 'f1_samples', 'f1_weighted', 'jaccard', 
             'jaccard_macro', 'jaccard_micro', 'jaccard_samples', 'jaccard_weighted'])
  min_features_to_select = Int: a quantidade mínima de features a escolher

  Retorno:
  ax = Gráfico contendo a avaliação por quantidade de features
  colunas_a_manter = Dataframe contendo o rfecv.support_ e as colunas de X 


  """
  min_features_to_select = min_features_to_select  # Minimum number of features to consider
  rfecv = RFECV(estimator=modelo, step=1, cv=StratifiedKFold(2),
                scoring= scorer,
              min_features_to_select=min_features_to_select)
  rfecv.fit(X, y)

  print("Optimal number of features : %d" % rfecv.n_features_)

  # Plot number of features VS. cross-validation scores
  plt.figure()
  plt.xlabel("Number of features selected")
  plt.ylabel("Cross validation score (nb of correct classifications)")
  ax = plt.plot(range(min_features_to_select,
               len(rfecv.grid_scores_) + min_features_to_select),
         rfecv.grid_scores_)
  plt.show()

  colunas_a_manter = pd.DataFrame({'coluna':X.columns,
              'bool': rfecv.support_})
  return ax, colunas_a_manter

def plota_correlacao(corr, annot):
  """
  Plota um Heatmap de correlação:
  Input:
  corr = Dataframe de correlação: normalmente representado pela função df.corr()
  annot = bool: Se quer que o heatmap venha com os valores ou não
  """
  plt.figure(figsize=(20,20))
  # Criando uma máscara para mostrar apenas os dados únicos do mapa de calor
  mask = np.triu(np.ones_like(corr, dtype=np.bool))
  heatmap = sns.heatmap(corr, mask=mask, vmin=-1, vmax=1, annot=annot, cmap='Blues', fmt = '.2f')
def cria_grafico_feature_importance(model,data):
  """
  Cria um gráfico com a importância das features num modelo de ensemble.
  Input:
  model: Modelo de Machine Learning Treinado, que contenha entre seus atributos o feature_importances_
  data: os dados usados para ajustar o modelo. (X)

  """
  importances = pd.Series(data=model.feature_importances_, index=data.feature_names)
  sns.barplot(x=importances, y=importances.index, orient='h').set_title('Importância de cada feature')

```

### **LIMPEZA DE DADOS**

Agora que fizemos o tratamento de nulos, vamos retirar os dados que não poderemos usar na nossa análise, conforme as diretrizes do Sírio-Libanês. Ou seja, precisamos retirar os dados relativos ao período em que os pacientes estiveram internados na UTI. No entanto, não podemos perder a informação de que este paciente foi ou não admitido na UTI em algum momento de sua visita, motivo pelo qual faremos uma manipulação da coluna ICU para manter a informação sem que usemos os dados deste período.

No primeiro momento, temos que retirar todos os dados dos pacientes que foram admitidos na UTI logo no início de sua visita:

```
#vamos agrupar por paciente para ver quem foi para a UTI nas primeiras duas horas
pacientes_icu = dados_sem_nulos.query('WINDOW=="0-2" and ICU==1')['PATIENT_VISIT_IDENTIFIER'].values
pacientes_icu
```

```
#retirando os clientes que foram para a UTI nas primeiras duas horas
dados_limpos = dados_sem_nulos.query('PATIENT_VISIT_IDENTIFIER not in @pacientes_icu')
```

Uma vez excluídos estes pacientes, vamos agora aplicar uma função que aplica na primeira janela de tempo a informação de admissão na UTI em qualquer momento de sua visita ao hospital. Desta forma, temos a certeza que os pacientes com status de ICU igual a 1 nesta janela foram admitidos em algum momento na UTI, mas não são os pacientes que foram admitidos na UTI nesta janela de tempo, pois excluímos os dados anteriormente

```
#confirmando que não há pacientes da UTI nas duas primeiras horas no novo dataset
dados_limpos.query('WINDOW=="0-2" and ICU==1')
```

```
#vamos unificar por visita do paciente, colocando na informação da primeira janela se ele foi ou não para a UTI
dados_unificados = dados_limpos.groupby('PATIENT_VISIT_IDENTIFIER').apply(prepara_janela).set_index('PATIENT_VISIT_IDENTIFIER').reset_index()
```

Sua documentação pode ser encontrada [aqui](https://pandas-profiling.github.io/pandas-profiling/docs/master/rtd/). O Pandas Profiling traz as seguintes informações:

* **Inferência de tipos** = Detecta os tipos das colunas num dataframe
* **Informações essenciais** = Tipo da coluna, valores únicos, valores faltantes
* **Estatísticas de quartis** = valor mínimo, primeiro quartil, mediana, terceiro quartil, máximo, amplitude dos dados e amplitude interquartis
* **Estatísticas descritivas** = tais como média, moda, desvio padrão, soma, mediana, desvio absoluto, coeficiente de variância, curtose, skewness(viés)
* **Valores mais frequentes**
* **Histogramas**
* **Correlações**  = Destacando ainda as variáveis com alta correlação, usando os métodos de Spearman, Pearson e Kendall
* **Dados Faltantes** = Matrix, contagem, mapa de calor e dendograma(diagrama que mostra a relação hierárquica entre as variáveis)
* **Linhas Duplicadas**
* **Análises de Texto** = aprende sobre as categorias (letra maíscula, espaços), tipo de script (latino, cirílico) e blocos (ASCII) de dados textuais

### **SELEÇÃO DE VARIÁVEIS**

A segunda fase da nossa análise exploratória é a escolha das variáveis que vamos utilizar para treinar nosso modelo de aprendizado de máquina. Para isso, vamos pegar nossos insights retirados dos relatórios do Pandas Profiling e SweetViz e fazer algumas visualizações usando métodos mais tradicionais, bem como tentar reduzir o número  de variáveis para o treinamento do modelo, para que ele não gaste recursos do hospital desnecessariamente. 

Para tanto, vamos partir do dataframe "dados_unificados" que contém todas as variáveis originais, e vamos aplicar diversas técnicas de seleção de variáveis até chegarmos às escolhidas para treinar nosso modelo.

Vimos que nossa abordagem inicial de excluir as medidas calculadas, embora na maior parte tenha se provado útil, exclui alguns dados que podem ser interessantes para o modelo (como as medidas de pressão arterial, taxa de batimentos cardíacos etc). Portanto vamos mudar nossa abordagem.

Dentre outras abordagens, vamos testar a hipótese de que podemos dispensar os resultados dos exames de sangue para nosso modelo, mantendo as features o mais simples possível, baseadas em testes mais simples e menos custosos para o hospital.

#### **EXCLUINDO AS VARIÁVEIS COM VALORES ÚNICOS**

```
#jogando numa nova variáveis:
#criando uma cópia do dataset
dados_sem_valores_unicos = dados_unificados.copy()
#para cada coluna cuja função unique retornar apenas um valor, vamos realizar o drop e substituir o dataset
for col in dados_sem_valores_unicos.columns:
    if len(dados_sem_valores_unicos[col].unique()) == 1:
        dados_sem_valores_unicos.drop(col,inplace=True,axis=1)
```

```
print(f'A função de redução de valores únicos reduziu em {dados_unificados.shape[1] -dados_sem_valores_unicos.shape[1]} colunas nosso dataset de estudo, e ele contém agora {dados_sem_valores_unicos.shape[1]} colunas')
```

#### **TRANSFORMANDO A COLUNA COM VALORES 100% DISTINTOS EM INDEX**

```
#determinando se há mais de uma coluna com valores distintos
distintos = []
for col in dados_sem_valores_unicos.columns:
  if len(dados_sem_valores_unicos[col].unique()) == dados_sem_valores_unicos.shape[0]:
    distintos.append(col)
distintos
```

```
#transformando a coluna em índice
dados_sem_valores_unicos.set_index('PATIENT_VISIT_IDENTIFIER', inplace=True)
dados_sem_valores_unicos.head()
```

#### **EXCLUINDO COLUNAS COM VALORES IDÊNTICOS ENTRE SI**

Para isso vamos usar a função drop_duplicates do pandas. Como ela exclui linhas ao invés de colunas, nós vamos primeiro transpor o dataframe para que os valores se transformem em linhas, e depois de excluir as linhas iguais, vamos voltar o dataframe ao estado original. #### **RODANDO UM ALGORITMO DE ESCOLHA DE FEATURES COM BASE NUM MODELO**

De acordo com este [artigo](https://operdata.com.br/blog/analise-de-componentes-principais/):

"*A Análise de Componentes Principais ou PCA (Principal Component Analysis) é uma técnica de análise multivariada que pode ser usada para analisar inter-relações entre um grande número de variáveis e explicar essas variáveis em termos de suas dimensões inerentes (Componentes).*

*O objetivo é encontrar um meio de condensar a informação contida em várias variáveis originais em um conjunto menor de variáveis estatísticas (componentes) com uma perda mínima de informação.*

*O número de componentes principais se torna o número de variáveis consideradas na análise, mas geralmente as primeiras componentes são as mais importantes já que explicam a maior parte da variação total.*

*As componentes principais em geral são extraídas via matriz de covariância, mas também podem ser extraídas via matriz de correlação.* "

Já a Eliminação Recursiva de Atributos (RFE), segundo a documentação do próprio [Scikit-Learn](https://scikit-learn.org/stable/modules/feature_selection.html#rfe):

"*Dado um estimador externo que aplica pesos às variáveis , o objetivo do RFE é selecionar variáveis considerando recursivamente conjuntos cada vez menores de variáveis*

*Na primeira vez que ele roda o estimador usando todas as variáveis possíveis e obtendo a resposta de "importância da variável" de cada atributo, ele retira as variáveis com menos importância, até que o número definido de variáveis seja selecionado*"

Neste caso, usamos a análise de componentes principais para determinar quantas variáveis explicam 95% dos dados, e após usamos este número para escolhermos nossas variáveis.

Vamos fazer o RFE usando dois métodos. Um com Validação Cruzada, para que ele mesmo escolha a quantidade de features usando o método StratifiedKFold:

```
#Separando as variáveis explicativas da variável alvo
X, y = dados_sirio_libanes.drop(['ICU'], axis=1), dados_sirio_libanes['ICU'].astype('int')
# Criando o modelo
rf = RandomForestClassifier()
```

```
from sklearn.utils.multiclass import type_of_target
type_of_target(y)
```

```
#vamos escolher como métrica o f1 e o modelo o Random Forest, e pedir que ele escolha pelo menos 2 features para o modelo
ax, colunas = selecao_features_por_modelo(X,y,rf,'f1', 2)
```

```
colunas_rfecv = colunas.query('bool == True').coluna.values
features_rfecv = list(colunas_rfecv)
```

Vamos também fazer uma nova escolha de features, desta vez usando o PCA (Análise de Componente Principal) e o RFE sem a validação cruzada

```
#CONFERINDO O PCA
from sklearn.decomposition import PCA
pca = PCA(0.95).fit(X)
pca.n_components_
```

```
#importando a biblioteca e buscando a ajuda do RFE
from sklearn.feature_selection import RFE
rfe= RFE(rf,pca.n_components_, step=1) #definindo 20 features de acordo com a resposta do PCA
#fazendo a seleção de features usando o RFE
rfe.fit(X, y)
#criando um Dataframe para as colunas a manter
colunas_a_manter = pd.DataFrame({'coluna':X.columns,
              'bool': rfe.support_})
```

```
#definindo as features
colunas_teste = colunas_a_manter.loc[colunas_a_manter['bool'] == True]
colunas_teste
```

```
#criando uma lista das variáveis
features_rfe = colunas_teste.coluna.values
features_rfe = list(features_rfe)
```
